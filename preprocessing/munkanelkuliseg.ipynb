```python
from repository.csv_parser import parse_package
import seaborn as sns

countyData = parse_package('../teir_adatok_2016/megye/')

munkanelkuliek = countyData[0]
```

```python
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=munkanelkuliek, kind="bar",
    x='title', y='A',
    ci="sd", palette="dark", alpha=.6, height=16
)
g.despine(left=True)
#g.set_axis_labels("", "Külföldiek vendéglátó egységekben")

g.savefig("output2.png")
```
