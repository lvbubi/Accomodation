```python
import visualization.diagrams as ds
import pandas as pd
import repository.accomodation_repository as repository

dataFrame = repository.loadTable()

allNights = dataFrame.foreign + dataFrame.domestic
#foreignStat = dataFrame.foreign / allNights

diagramModel = pd.concat([dataFrame.county, allNights], axis=1, sort=False)
```

```python
diagramModel.county = diagramModel.county.apply(lambda x: x.split("-")[0] if "-" in x else x)
diagramModel = diagramModel.sort_values(0, ascending=False)
ds.plotHistogram(diagramModel)
```
